```python
# Identify the overlapping time period for calculation
# The df_business_applications_ts starts in 2025-07-01 and only has a few data points.
# It seems there's an issue with the range of df_business_applications_ts data.
# Based on the plot, df_business_applications_ts only has data for 2025-07.
# Let's recalculate the time index for df_business_applications_ts based on the available data.
# From the head of the dataframe, the columns are Monthly_Jul_2025, Jun_2025, etc.
# This indicates the columns are monthly data points, not rows.
# Let's re-examine df_business_applications after the header fix and melting step.
# It seems the previous melting attempt and subsequent steps with df_business_applications_ts assumed the rows were time points, which is incorrect based on the column headers in the raw data.

# Let's re-process df_business_applications to get a proper time series DataFrame.
# Based on the cleaned df_business_applications, the columns represent months.
# We need to melt this dataframe to get a time series.

# Identify the month columns and application type column
app_type_col = 'Application_Type'
month_cols = [col for col in df_business_applications.columns if col != app_type_col and not col.startswith('Percent_Change')]

# Melt the dataframe to long format
df_business_applications_melted = df_business_applications.melt(
    id_vars=[app_type_col],
    value_vars=month_cols,
    var_name='Month_Year',
    value_name='Value'
)

# Create a datetime index from the 'Month_Year' column
def parse_month_year(month_year_str):
    parts = month_year_str.split('_')
    if len(parts) >= 2:
        month_abbr = parts[1]
        year = parts[2] if len(parts) > 2 and parts[2].isdigit() else None
        if month_abbr in month_map and year:
            return f"{year}-{month_map[month_abbr]:02d}-01"
    return None

df_business_applications_melted['date'] = df_business_applications_melted['Month_Year'].apply(parse_month_year)
df_business_applications_melted['date'] = pd.to_datetime(df_business_applications_melted['date'])

# Drop rows where date could not be parsed
df_business_applications_melted.dropna(subset=['date'], inplace=True)

# Set the date as index and pivot to get application types as columns
df_business_applications_ts_corrected = df_business_applications_melted.pivot_table(
    index='date',
    columns='Application_Type',
    values='Value',
    aggfunc='sum' # Use sum in case there are multiple entries for the same date/app_type
)

# Display the corrected time series dataframe
print("Corrected Business Applications Time Series:")
display(df_business_applications_ts_corrected.head())
display(df_business_applications_ts_corrected.tail())

# Now identify the overlapping time period between the BFS data and the corrected business applications data
# Find the minimum start date and maximum end date across the dataframes with datetime indices
start_date_bfs_us = df_bfs_monthly_agg.index.min() if not df_bfs_monthly_agg.empty else None
end_date_bfs_us = df_bfs_monthly_agg.index.max() if not df_bfs_monthly_agg.empty else None

start_date_bfs_pr = df_bfs_monthly_pr_agg.index.min() if not df_bfs_monthly_pr_agg.empty else None
end_date_bfs_pr = df_bfs_monthly_pr_agg.index.max() if not df_bfs_monthly_pr_agg.empty else None

start_date_ba = df_business_applications_ts_corrected.index.min() if not df_business_applications_ts_corrected.empty else None
end_date_ba = df_business_applications_ts_corrected.index.max() if not df_business_applications_ts_corrected.empty else None

# Determine the common start and end dates
common_start_date = max([dt for dt in [start_date_bfs_us, start_date_bfs_pr, start_date_ba] if dt is not None]) if any([start_date_bfs_us, start_date_bfs_pr, start_date_ba]) else None
common_end_date = min([dt for dt in [end_date_bfs_us, end_date_bfs_pr, end_date_ba] if dt is not None]) if any([end_date_bfs_us, end_date_bfs_pr, end_date_ba]) else None

print(f"\nCommon Time Period: {common_start_date} to {common_end_date}")

if common_start_date and common_end_date and common_start_date <= common_end_date:
    # Filter dataframes to the overlapping period
    bfs_us_overlap = df_bfs_monthly_agg.loc[common_start_date:common_end_date]
    bfs_pr_overlap = df_bfs_monthly_pr_agg.loc[common_start_date:common_end_date]
    ba_overlap = df_business_applications_ts_corrected.loc[common_start_date:common_end_date]

    print("\nDescriptive Statistics for BFS US (Overlapping Period):")
    display(bfs_us_overlap.describe())

    print("\nDescriptive Statistics for BFS PR (Overlapping Period):")
    display(bfs_pr_overlap.describe())

    print("\nDescriptive Statistics for Business Applications (Overlapping Period):")
    display(ba_overlap.describe())
else:
    print("\nNo overlapping time period found for calculating descriptive statistics.")
```
