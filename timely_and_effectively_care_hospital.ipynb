```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def load_and_clean_data(file_path):
    """
    Loads the healthcare data from a CSV file, converts 'Not Available' in the
    'Score' column to NaN, creates a 'Score_numeric' column, and drops the
    'Sample' and 'Footnote' columns.

    Args:
        file_path (str): The path to the CSV file.

    Returns:
        pandas.DataFrame: The cleaned DataFrame with 'Score_numeric' and
                          without 'Sample' and 'Footnote' columns.
    """
    try:
        df = pd.read_csv(file_path)
        df['Score_numeric'] = pd.to_numeric(df['Score'], errors='coerce')
        df_cleaned = df.drop(columns=['Sample', 'Footnote'])
        print("Data loaded and initially cleaned successfully.")
        return df_cleaned
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        return None
    except Exception as e:
        print(f"An error occurred during data loading and cleaning: {e}")
        return None

def explore_data(df):
    """
    Displays the first 5 rows, information about the DataFrame (including
    column types and non-null counts), and the count of missing values per column.

    Args:
        df (pandas.DataFrame): The DataFrame to explore.
    """
    if df is not None:
        print("First 5 rows of the DataFrame:")
        display(df.head())
        print("\nDataFrame Information:")
        display(df.info())
        print("\nMissing values per column:")
        display(df.isnull().sum())
    else:
        print("Cannot explore data: DataFrame is None.")

def plot_numerical_distribution(df, numerical_col, title='Distribution', xlabel='Value', ylabel='Frequency'):
    """
    Creates and displays a histogram or density plot of the specified numerical column,
    handling NaN values.

    Args:
        df (pandas.DataFrame): The DataFrame containing the numerical column.
        numerical_col (str): The name of the numerical column to plot.
        title (str, optional): The title of the plot. Defaults to 'Distribution'.
        xlabel (str, optional): The label for the x-axis. Defaults to 'Value'.
        ylabel (str, optional): The label for the y-axis. Defaults to 'Frequency'.
    """
    if df is not None and numerical_col in df.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[numerical_col].dropna(), kde=True)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.show()
    elif df is None:
        print("Cannot plot numerical distribution: DataFrame is None.")
    else:
        print(f"Cannot plot numerical distribution: Column '{numerical_col}' not found.")


def analyze_categorical_column(df, categorical_col):
    """
    Analyzes a categorical column by printing the number of unique values,
    the first 20 unique values, and the value counts (top 10).

    Args:
        df (pandas.DataFrame): The DataFrame containing the categorical column.
        categorical_col (str): The name of the categorical column to analyze.
    """
    if df is not None and categorical_col in df.columns:
        unique_values = df[categorical_col].unique()
        print(f"Number of unique values in '{categorical_col}':", len(unique_values))
        print(f"First 20 unique values in '{categorical_col}':", unique_values[:20])

        value_counts = df[categorical_col].value_counts()
        print(f"\nValue counts for '{categorical_col}' (Top 10):\n", value_counts.head(10))
    elif df is None:
        print("Cannot analyze categorical column: DataFrame is None.")
    else:
        print(f"Cannot analyze categorical column: Column '{categorical_col}' not found.")

def plot_categorical_numerical_relationship(df, categorical_col, numerical_col, threshold=50, figsize=(15, 8), title='Distribution of Scores by Category', xlabel=None, ylabel='Score'):
    """
    Creates and displays a box plot of a numerical column across categories
    in a categorical column, filtering categories based on a threshold of
    minimum data points.

    Args:
        df (pandas.DataFrame): The DataFrame containing the columns.
        categorical_col (str): The name of the categorical column.
        numerical_col (str): The name of the numerical column.
        threshold (int, optional): Minimum number of data points for a category
                                   to be included in the plot. Defaults to 50.
        figsize (tuple, optional): The size of the figure. Defaults to (15, 8).
        title (str, optional): The title of the plot. Defaults to 'Distribution of Scores by Category'.
        xlabel (str, optional): The label for the x-axis. Defaults to the categorical column name.
        ylabel (str, optional): The label for the y-axis. Defaults to 'Score'.
    """
    if df is not None and categorical_col in df.columns and numerical_col in df.columns:
        counts = df[categorical_col].value_counts()
        categories_to_plot = counts[counts > threshold].index.tolist()
        df_filtered = df[df[categorical_col].isin(categories_to_plot)].copy()

        plt.figure(figsize=figsize)
        sns.boxplot(x=categorical_col, y=numerical_col, data=df_filtered)
        plt.title(title)
        plt.xlabel(xlabel if xlabel is not None else categorical_col)
        plt.ylabel(ylabel)
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    elif df is None:
        print("Cannot plot relationship: DataFrame is None.")
    else:
        print(f"Cannot plot relationship: One or more columns ('{categorical_col}', '{numerical_col}') not found.")


def perform_correlation_analysis(df, numerical_cols):
    """
    Performs correlation analysis on specified numerical columns, handling
    missing values by dropping rows, and displays a heatmap of the correlation matrix.

    Args:
        df (pandas.DataFrame): The DataFrame containing the numerical columns.
        numerical_cols (list): A list of numerical column names for correlation analysis.

    Returns:
        pandas.DataFrame: The correlation matrix. Returns None if analysis cannot be performed.
    """
    if df is not None and all(col in df.columns for col in numerical_cols):
        df_numerical = df[numerical_cols].copy()
        df_numerical_cleaned = df_numerical.dropna()

        if not df_numerical_cleaned.empty:
            correlation_matrix = df_numerical_cleaned.corr()

            plt.figure(figsize=(8, 6))
            sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
            plt.title('Correlation Matrix of Numerical Variables')
            plt.show()
            return correlation_matrix
        else:
            print("No non-null data points for correlation analysis after dropping NaNs.")
            return None
    elif df is None:
        print("Cannot perform correlation analysis: DataFrame is None.")
        return None
    else:
        missing_cols = [col for col in numerical_cols if col not in df.columns]
        print(f"Cannot perform correlation analysis: Columns not found - {missing_cols}")
        return None

def perform_state_analysis(df, measure_name, threshold=10, figsize=(14, 7)):
    """
    Analyzes and visualizes the average numerical score per state for a
    specific measure, filtering states based on a minimum number of data points.

    Args:
        df (pandas.DataFrame): The DataFrame containing the data.
        measure_name (str): The name of the measure to analyze.
        threshold (int, optional): Minimum number of data points for a state
                                   to be included in the analysis. Defaults to 10.
        figsize (tuple, optional): The size of the figure. Defaults to (14, 7).

    Returns:
        pandas.Series: A pandas Series with average scores per state, sorted
                       in descending order. Returns None if analysis cannot be performed.
    """
    if df is not None and 'Measure Name' in df.columns and 'State' in df.columns and 'Score_numeric' in df.columns:
        df_measure = df[df['Measure Name'] == measure_name].copy()
        df_measure = df_measure.dropna(subset=['Score_numeric'])

        if not df_measure.empty:
            average_score_per_state = df_measure.groupby('State')['Score_numeric'].mean().sort_values(ascending=False)

            # Filter states based on threshold
            state_counts = df_measure['State'].value_counts()
            states_to_include = state_counts[state_counts >= threshold].index.tolist()
            df_filtered_states = df_measure[df_measure['State'].isin(states_to_include)].copy()

            if not df_filtered_states.empty:
                average_score_per_state_filtered = df_filtered_states.groupby('State')['Score_numeric'].mean().sort_values(ascending=False)

                plt.figure(figsize=figsize)
                sns.barplot(x='State', y='Score_numeric', data=df_filtered_states, estimator=np.mean, errorbar=None, order=average_score_per_state_filtered.index)
                plt.title(f'Average Score for "{measure_name}" by State (States with >= {threshold} data points)')
                plt.xlabel('State')
                plt.ylabel('Average Score')
                plt.xticks(rotation=45, ha='right')
                plt.tight_layout()
                plt.show()
                return average_score_per_state
            else:
                print(f"No states found with at least {threshold} data points for measure '{measure_name}'.")
                return None
        else:
            print(f"No non-null scores found for measure '{measure_name}'.")
            return None
    elif df is None:
        print("Cannot perform state analysis: DataFrame is None.")
        return None
    else:
        missing_cols = [col for col in ['Measure Name', 'State', 'Score_numeric'] if col not in df.columns]
        print(f"Cannot perform state analysis: Columns not found - {missing_cols}")
        return None


def summarize_analysis_findings(summary_text):
    """
    Prints a structured summary of the key findings from the analysis.

    Args:
        summary_text (str): A string containing the summary of findings.
    """
    print("\n--- Summary of Analysis Findings ---")
    print(summary_text)
    print("--- End of Summary ---")
```
